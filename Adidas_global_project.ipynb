```python
#fill the null value in the object datatype column with mode of the column
def fill_object_null_with_mode(df):
  object_cols=df.select_dtypes(include=['object']).columns

  for col in object_cols:
    mode_value=df[col].mode()[0]
    df[col]=df[col].fillna(mode_value)

  return df

df=fill_object_null_with_mode(df)
```

```python
df.describe()
```

```python
def fill_null_value_int_float_col(df):
  numeric_col=df.select_dtypes(include=[int,float]).columns

  for col in numeric_col:
    mean_value=df[col].mean()
    median_value=df[col].median()

    difference=abs(mean_value-median_value)
    if difference>3:
      fill_value=median_value

    else:
      fill_value=mean_value

    df[col]=df[col].fillna(fill_value)

  return df




```

```python
df=fill_null_value_int_float_col(df)
```
